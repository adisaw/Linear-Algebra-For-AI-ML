```python
from matplotlib.image import imread
from matplotlib import pyplot

import cv2
import numpy as np

image = "dog.jpg"
img=imread(image)
pyplot.imshow(img)
```

```python
D= img
D= cv2.cvtColor(D,cv2.COLOR_BGR2GRAY)
pyplot.imshow(D, cmap="gray",vmin=0,vmax=255)
```

```python
print(D)
print(D.shape)
```

```python
p=5
val=int((p-1)/2)
valx=val+D.shape[0]
valy=val+D.shape[1]
array=np.zeros(shape=(D.shape[0]+p-1,D.shape[1]+p-1))
array2=np.zeros(shape=(D.shape[0]+p-1,D.shape[1]+p-1))
array2[val:valx,val:valy]=D
#print(array2.shape)
#print(array2)
B=np.zeros(shape=(p,p))
for i in range(p):
    for j in range(p):
        B[i][j]=1/(p*p)


for r in range(array.shape[0]):
    for s in range(array.shape[1]):
        summ=0
        k=r-val
        
        for i in range(p):
            l=s-val
            for j in range(p):
                if(k>=0 and k<array.shape[0] and l>=0 and l<array.shape[1]):
                    summ=summ+B[i][j]*array2[k][l]
                l=l+1
            k=k+1
        array[r][s]=summ

#print(array)

                    
pyplot.imshow(array, cmap="gray",vmin=0,vmax=255)
```
