```python
import math
import matplotlib.pyplot as mp
import numpy as np
from mpl_toolkits import mplot3d
x=2*math.pi
x=x/1000
print(x)
```

```python
arr=np.zeros((1000,2))
for i in range(1000):
    arr[i][0]=math.cos(i*x)
    arr[i][1]=math.sin(i*x)
mp.axis('equal')
mp.scatter(arr[:,0],arr[:,1])
```

```python
arr2=np.zeros((1000,3))
#a) 
mat1=[[-1/math.sqrt(2),0],[0,-1/math.sqrt(2)],[-1,1]]
for i in range(1000):
    t=np.matmul(mat1,arr[i,:].T)
    arr2[i,:]=t.T

ax = mp.axes(projection='3d')
#ax.axes('equal')
ax.scatter3D(arr2[:,0],arr2[:,1],arr2[:,2])
```

```python
mp.axis('equal')
mp.scatter(arr2[:,0],arr2[:,1])
```

```python
print('Condition Number of 9a matrix',np.linalg.cond(mat1))
```

```python
import random
arr3=np.zeros((1000,3))
def random_float():
    return random.random()*(20)+-10

for i in range(1000):
    arr3[i,0]=random_float()
    arr3[i,1]=random_float()
    arr3[i,2]=random_float()
    arr3[i,:]=arr3[i,:]/np.linalg.norm(arr3[i,:])
    #print(np.linalg.norm(arr3[i,:]))

ax = mp.axes(projection='3d')
ax.set_aspect('equal')
ax.scatter3D(arr3[:,0],arr3[:,1],arr3[:,2])

```

```python
#b)
mat2=[[-2,1,2],[0,2,0]]
arr4=np.zeros((1000,2))
for i in range(1000):
    x=np.matmul(mat2,arr3[i,:].T)
    arr4[i,:]=x.T

mp.axis('equal')
mp.scatter(arr4[:,0],arr4[:,1])
```

```python
print('Condition Number of 9b matrix',np.linalg.cond(mat2))
```
